````python
import random

"""Monopoly.py"""
class MonopolyProperty:
    
    
    """METHODS TO BE IMPLEMENTED:"""
    def __init__(self, n,p,r):
        """
        Initializes the MonopolyProperty object setting 
        its  name to n, its price to p and its rent to r"""
        self.__name = n
        self.__price = p
        self.__rent = r
        self.__nextProperty = None
        self.__prevProperty = None
    

    def getName(self): 
        """returns the name of the property"""
        return self.__name

    def getPrice(self): 
        """returns the buying price of the property"""        
        return self.__price
    
    def getRent(self): 
        """returns the rent price of the property"""           
        return self.__rent
    
    def setNext(self, other): 
        """gets another MonopolyProperty other and sets it as the next property"""
        self.__nextProperty = other
        
    def setPrevious(self, other): 
        """gets another MonopolyProperty other and sets it as the previous property"""
        self.__prevProperty = other
        
    def getNext(self): 
        """returns the next property"""
        return self.__nextProperty

    def getPrevious(self): 
        """returns the previous property"""
        return self.__prevProperty    
    
    def moveFWDn(self, n): 
        """with n positive integer, returns the MonopolyProperty that is n properties 
        after the current property (hint: move to the next property n times)."""
        r = self
        for i in range(0,n):
            r = r.getNext()
        
        return r

    def moveFWDnew(self, n, prop): 
        """with n positive integer, returns the MonopolyProperty that is n properties 
        after the current property (hint: move to the next property n times)."""
        r = self
        crossed = False
        for i in range(0,n):
            r = r.getNext()
            if r == prop:
                crossed = True
        
        return (r,crossed)
    
    def moveREVn(self, n): 
        """with n positive integer, returns the MonopolyProperty that is n properties 
        before the current property (hint: move to the previous property n times)."""            
        r = self
        for i in range(0,n):
            r = r.getPrevious()
        
        return r
            

class MonopolyBoard:
    """ADD ATTRIBUTES HERE"""
    
    
    """METHODS TO BE IMPLEMENTED"""
    def __init__(self, startProperty):
        """initializes the board with the MonopolyProperty startProperty as start
        setting also its previous and next pointers to itself
        """
        startProperty.setPrevious(startProperty)
        startProperty.setNext(startProperty)
        self.__start = startProperty

    def getStart(self): 
        """returns the start **MonopolyProperty**"""
        return self.__start
        
    def addProperty(self, newProperty): 
        """adds the MonopolyProperty **newProperty** at the end of the circular linked list (hint: 
           it has to become the previous element of start -- remember to update correctly all 
           the previous and next pointers)"""
        if type(newProperty == MonopolyProperty):
            P = self.__start.getPrevious()
            P.setNext(newProperty)
            newProperty.setPrevious(P)
            newProperty.setNext(self.__start)
            self.__start.setPrevious(newProperty)
            
            
    def __str__(self):
        outSTR = ""
        curProp = self.__start
        cnt = 1
        outSTR = "[{}] {}\tprice: {} rent: {}\n".format(cnt, 
                                                    curProp.getName(),
                                                    curProp.getPrice(),
                                                    curProp.getRent())
        curProp = curProp.getNext()
        cnt += 1
        
        while curProp != self.__start:
            outSTR += "[{}] {}\tprice: {} rent: {}\n".format(cnt, 
                                            curProp.getName(),
                                            curProp.getPrice(),
                                            curProp.getRent())
            cnt += 1
            curProp = curProp.getNext()
        
        return outSTR
        

if __name__ == "__main__":
    Monopoly = None
    inF = open("mnply.csv", "r")
    
    for line in inF:
        line = line.strip()
        if not line.startswith("Name"):
            els = line.split(",")
            if els[0] == "Go":
                tile = MonopolyProperty(els[0], int(els[2]), int(els[3]))
                Monopoly = MonopolyBoard(tile)
            else:
                tile = MonopolyProperty(els[0], int(els[2]), int(els[3]))
                Monopoly.addProperty(tile)
    inF.close()
    print("THE BOARD:")
    print(Monopoly)
    
    #let's simulate a game of 20 rounds, with an initial pocket of $150. 
    player = [Monopoly.getStart(), 150]
    
    for i in range(20):
        dice = random.randint(1,6)
        player[0] = player[0].moveFWDn(dice)
        player[1] -= player[0].getRent()
        

        print("[rnd:{}] Dice: {} Player is now on {} with ${}".format(i+1, 
                                                                       dice,
                                                            player[0].getName(),
                                                            player[1]))
        
        if player[1] <= 0:
            print("Player is bankrupt!")
            break
            

    print("\n\nANOTHER round (with modidifed moveFWD)")
    player = [Monopoly.getStart(), 150]
    for i in range(20):
        dice = random.randint(1,6)
        move = player[0].moveFWDnew(dice, Monopoly.getStart())
        player[0] = move[0]#player[0].moveFWDn(dice)
        player[1] -= player[0].getRent()
        
        if move[1] == True:
            print("Lap finished, added $100")
            player[1] += 100
        print("[rnd:{}] Dice: {} Player is now on {} with ${}".format(i+1, 
                                                                       dice,
                                                            player[0].getName(),
                                                            player[1]))
        
        if player[1] <= 0:
            print("Player is bankrupt!")
            break
        



````
