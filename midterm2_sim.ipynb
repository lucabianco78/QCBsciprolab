````python
"""exercise1.py"""

def F(n):
    if n < 0:
        print("n must be > 0")
        return None
    if n < 2:
        return 2*n
    else:
        n_minus2 = 0
        n_minus1 = 2
        for i in range(2,n+1):
            newVal = i *n_minus1 - n_minus2
            n_minus2 = n_minus1
            n_minus1 = newVal
        return n_minus1
    
for i in range(10):
    print("F({}) = {}".format(i,F(i)))
print("\n")
i=1320
print("F({}) = {}".format(i,F(i)))
````

````python
%reset -f
from collections import deque
import math

"""exercise2.py"""

class supermarket:
    #0. the initializer
    def __init__(self, n):
        self.cashCounters = []
        self.totalCustomers = 0
        self.waitingCustomers = 0
        for i in range(n):
            self.cashCounters.append(deque())
        
    #1. add customerToCounter
    def addCustomerToCounter(self, counterID, customerID):
        if counterID < len(self.cashCounters):
            self.cashCounters[counterID].append(customerID)
            self.totalCustomers += 1
            self.waitingCustomers += 1
        
    #2. addCustomer(self, customerID)
    def addCustomer(self, customerID):
        counterID = -1
        m = math.inf
        for i in range(len(self.cashCounters)):
            L = len(self.cashCounters[i])
            if L < m:
                counterID = i
                m = L
        if counterID != -1:
            self.cashCounters[counterID].append(customerID)
            self.totalCustomers += 1
            self.waitingCustomers += 1
            
    #3. removeCustomer(counterID)
    def removeCustomer(self,counterID):
        customer = None
        if counterID < len(self.cashCounters):
            if len(self.cashCounters[counterID]) > 0:
                customer = self.cashCounters[counterID].popleft()
                self.waitingCustomers -= 1
            return customer
        
    #4. __str__()
    def __str__(self):
        outSTR = "{} total customers served\n".format(self.getTotalCustomers())
        outSTR +="{} customers still waiting\n".format(self.getWaitingCustomers())
        outSTR += "Current queues:\n"
        for i in range(len(self.cashCounters)):
            outSTR +="\tCounter_{}: {}\n".format(i, list(self.cashCounters[i]))
        return outSTR
        
    #5.  getTotalCustomers(self)
    def  getTotalCustomers(self):
        return self.totalCustomers
    
    #6.  getWaitingCustomers(self)
    def  getWaitingCustomers(self):
        return self.waitingCustomers

    
mySuper = supermarket(4)

for i in range(15):
    mySuper.addCustomer("c"+ str(i))
print(mySuper)
mySuper.addCustomerToCounter(0, "c16")
mySuper.addCustomerToCounter(0, "c17")
mySuper.addCustomerToCounter(0, "c18")
mySuper.addCustomerToCounter(1, "c19")
mySuper.addCustomerToCounter(2, "c20")
print(mySuper)
for i in range(12):
    c = mySuper.removeCustomer(i%4)
    print("Customer {} exiting {}".format(c, "Counter_"+str(i%4)))

print("")
print(mySuper)
mySuper.addCustomer("c21")
print(mySuper)



````
