def fun(L,i,j):
    if i==j:
        return 0
    else:
        m = (i+j)//2
        count = 0
        for x in L[i:m]:
          for y in L[m:j+1]:
             if x==y:
                count = count+1
        left = fun(L,i,m)
        right = fun(L,m+1,j)
        return left+right+count

````python
"""
NAME:
SURNAME:
Matricola: 

partA.py
"""
import pandas as pd
from matplotlib import pyplot as plt

def computestats(data):
    """
    gets the data structure loaded and prints the total 
    number of chromosomes, the total number of SNPs and 
    the total number of SNPs for each possible type 
    (i.e. TA, TC, TG, TT, AT,AC, AG, AA,...) and 
    the average read depth (with 3 decimal places of precision)
    """
    print("The file contains {} SNPs".format(data.shape[0]))
    
    countsByC = data['Chr'].value_counts()
    print("Belonging to {} chromosomes:".format(len(countsByC)))
    chrs = []
    for i in countsByC.index:
        chrs.append(i)
    chrs.sort()
    print("\n".join(chrs))
    
    data['type'] = data['Ref']+data['Alt']
    countsByType = data['type'].value_counts()
    for el in countsByType.index:
        print("Type {} counts {} SNPs".format(el, countsByType[el]))
        
    print("Average read depth: {:.3f} reads per site".format(data["ReadDepth"].mean()))
    
    
def chrHistogram(data,chromosome, start, end):
    """
    given a chromosome, a start postion and an end position,  
    plots an histogram with the number of SNPs for each type
    belonging to that chromosome having postion in 
    the interval [start, end]
    """
    chrFilt = data[data['Chr'] == chromosome]
    tmp = chrFilt[chrFilt['Pos'] >= start]
    filtSNPs = tmp[tmp['Pos'] <= end]
    countsByType = filtSNPs['type'].value_counts()
    countsByType.plot(kind = 'bar')
    plt.title('# SNPs of {} in [{},{}]'.format(chromosome, start,end))
    plt.show()
    pass
    


def readDepthStats(data, chromosome, start, end):
    """
    given the input data and chromosome, a start and end position, 
    the function computes  the average read depth of all the SNPs 
    belonging to that chromosome (and in the [start,end] interval) 
    and the average read depth for each 
    type of SNP (TA, TC, ...).
    """
    chrFilt = data[data['Chr'] == chromosome]
    tmp = chrFilt[chrFilt['Pos'] >= start]
    filtSNPs = tmp[tmp['Pos'] <= end]
    print("Global Avg. read depth for SNPs in {} [{},{}]: {:.3f}".format(chromosome,
                                                             start,
                                                             end,
                                                             filtSNPs['ReadDepth'].mean()))
    
    byType = filtSNPs.groupby('type')['ReadDepth'].mean()
    for el in byType.index:
        print("\tType: {} avg. read depth: {:.3f}".format(el,byType[el]))
    pass

dataFile = "file_samples/snp_data.tsv"

data = pd.read_csv(dataFile,
                     sep="\t",  header = 0)
    

computestats(data)
print("\n")
chrHistogram(data, "Chr01",500,4000)
chrHistogram(data, "Chr02",1,1250)
chrHistogram(data, "Chr03",1,50000)
readDepthStats(data, "Chr01", 500,6000)
print("\n")
readDepthStats(data, "Chr02",1, 2000)

    
    


````

````python
%reset -f 
"""
NAME:
SURNAME:
Matricola: 

partB.py
"""

from collections import deque

class BinaryTree:
    def __init__(self, value):
        self.__data = value
        self.__right = None
        self.__left = None
        self.__parent = None
    
        
    def getValue(self):
        return self.__data
        
    def getParent(self):
        return self.__parent
    
    #needed because we are using private attributes
    def setParent(self, tree):
        self.__parent = tree
    
    def getRight(self):
        return self.__right
    
    def getLeft(self):
        return self.__left
    
    def insertRight(self, tree):
        if self.__right == None:
            self.__right = tree
            tree.setParent(self)
    
                    
    def insertLeft(self, tree):
        if self.__left == None:
            self.__left = tree
            tree.setParent(self)
            
    def deleteRight(self):
        self.__right = None
        
    def deleteLeft(self):
        self.__left = None
        

def createBST(intList):
    BST = None
    if len(intList) > 0:
        BST = BinaryTree(intList[0])
        for el in intList[1:]:
            cur_el = BST
            alreadyPresent = False
            prev_el = None
            while cur_el != None:
                prev_el = cur_el
                cv = cur_el.getValue()
                if  cv > el:
                    cur_el = cur_el.getLeft()
                elif cv < el:
                    cur_el = cur_el.getRight()
                else:
                    #cv == el (el is already present)
                    #not allowed by rule c. skip it
                    alreadyPresent = True
                    #print("El {} already present".format(el))
                    break
                
                
            if not alreadyPresent:
                node = BinaryTree(el)
                node.setParent(prev_el)
                if prev_el.getValue() > el:
                    prev_el.insertLeft(node)
                else:
                    prev_el.insertRight(node)
                
        
    return BST

def findExtremes(BST):
    """
    returns a tuple with two elements that are the smaller and the 
    bigger element in the BST
    """
    root = BST
    minVal = None
    maxVal = None
    if root == None:
        return (minVal, maxVal)
    
    #Find the smallest element
    curNode = root
    while curNode.getLeft() != None:
        curNode = curNode.getLeft()
        
    minVal = curNode.getValue()
    
    #Find the biggest element
    curNode = root
    while curNode.getRight() != None:
        curNode = curNode.getRight()
    maxVal = curNode.getValue()
    
    return (minVal, maxVal)


def sumNodes(BST, N):
    """
    given a binary search tree BST and a node N, 
    returns the sum of all the elements that are in 
    the subtree rooted at N or None if N is not in BST
    """
    #find node N
    curNode = BST
    while curNode != None and curNode.getValue() != N:
        if N > curNode.getValue():
            curNode = curNode.getRight()
        else:
            curNode = curNode.getLeft()
            
    if curNode == None:
        return None
    else:
        res = curNode.getValue()
        subtree = deque()
        subtree.append(curNode.getRight())
        subtree.append(curNode.getLeft())
        while len(subtree) > 0:
            curNode = subtree.popleft()
            if curNode != None:
                res += curNode.getValue()
                subtree.append(curNode.getRight())
                subtree.append(curNode.getLeft())
        return res
        
    
def findClosestValue(BST,value):    
    """
    given an integer value the method should return 
    value if present in the BST,  or a list of all the 
    values that are closest to value (higher or lower)
    """
    #find node value
    dist = abs(BST.getValue() - value)
    #this will be filled in the first loop of while
    closestEls = []
    curNode = BST
    
    while curNode != None:
        curVal = curNode.getValue()
        if curVal == value:
            return value
        tmpD = abs(curVal - value)
        if tmpD < dist:
            closestEls = [curVal]
            dist = tmpD
        else:
            if tmpD == dist:
                closestEls.append(curVal)
        #move to next
        if value > curVal:
            curNode = curNode.getRight()
        else:
            curNode = curNode.getLeft()
    
    
    return closestEls
    
    
def plotGraph(tree):    
    #uses pygraphviz 
    G=pgv.AGraph(directed=True)
    
    levels = [tree]
    while len(levels) > 0:
        cur_n = levels.pop()
        if cur_n != None:
            G.add_node(cur_n.getValue(), color = 'black')
            r = cur_n.getRight()
            l = cur_n.getLeft()
            if l != None:
                G.add_node(l.getValue(), color = 'black')
                G.add_edge(cur_n.getValue(), l.getValue())
                levels.append(l)
            if r != None:
                G.add_node(r.getValue(), color = 'black')
                G.add_edge(cur_n.getValue(), r.getValue())
                levels.append(r)
    G.layout(prog='dot') # use dot
    #In the following, the path of the graph. Change
    #it to your liking
    G.draw('img/testbsttree.png')
    
    
if __name__ == "__main__":
    import pygraphviz as pgv
    intL = [9,25, 32, 31, 144,156, 44, 40, -1, 72, 11, 4, -7, 3, 1, 6]
    
    B = createBST(intL)
    r = findExtremes(B)
    print("Min value: {}\nMax value: {}".format(r[0],r[1]))
    plotGraph(B)
    print("Sum subtree rooted at 4: {}".format(sumNodes(B,4)))
    print("Sum subtree rooted at -1: {}".format(sumNodes(B,-1)))
    print("Sum subtree rooted at 25: {}".format(sumNodes(B,25)))
    print("Sum subtree rooted at 144: {}".format(sumNodes(B,144)))
    print("Sum subtree rooted at 71: {}".format(sumNodes(B,71)))
    
    print("Element(s) closest to 25: {}".format(findClosestValue(B,25)))
    print("Element(s) closest to 10: {}".format(findClosestValue(B,10)))
    print("Element(s) closest to 5: {}".format(findClosestValue(B,5)))
    print("Element(s) closest to 28: {}".format(findClosestValue(B,28)))
    print("Element(s) closest to 36: {}".format(findClosestValue(B,36)))
    
    
````
