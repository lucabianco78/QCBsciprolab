````python
"""
NAME:
SURNAME:
Matricola: 

exercise1.py
"""

import pandas as pd
import matplotlib.pyplot as plt



def loadTSV(datafile):
    """gets the filename of the tsv file and returns a pandas dataframe with all the loaded information"""
    df = pd.read_csv(datafile,
                     sep="\t", header = 0)
    return df
    
def computestats(data):
    """gets the data structure loaded with loadTSV and prints the total number of sequences, 
    the total number of LGs, and for each LG the total number of contigs, total size and total sum of gaps. 
    Finally, the total size of the genome should be printed (i.e. the sum of all the total sizes)"""
    print("Total number of sequences: {}".format(data.shape[0]))
    lgData = data["LG"].value_counts()
    lgs = list(lgData.index)
    lgs.sort()
    print("Total number of LGs: {}\n".format(len(lgs)))
    totSize = 0
    for el in lgs:
        tmp = data[data["LG"] == el]
        M = tmp["End"].max()
        gaps = M - tmp["Size"].sum()
        totSize += M
        print("{} has {} seq. (TOT size {:,} bps gaps: {:,} bps)".format(el, lgData[el], M, gaps))
    print("\nThe total size of the genome is {:,} bps".format(totSize))
        
        
     
def computeN50(data, lg = None):
    """given the input data, and an LG (if unspecified all data should be considered) and returns the N50. 
    If the LG does not exist a warning message should be returned."""
    tmp = data
    if lg != None:
        tmp = tmp[tmp["LG"] == lg]
    
    if tmp.shape[0] == 0:
        print("Warning cannot find information on {}".format(lg))
    else:
        sizes = list(tmp["Size"])
        sumS = tmp["Size"].sum()
        sizes.sort(reverse = True)
        i = 0
        cumSum = 0
        while i < len(sizes) and cumSum < sumS/2.0:
            cumSum += sizes[i]
            i += 1
        return sizes[i-1]
        

def plotN50(data,lg = None):
    """given the input data, an lg (optional, if unspecified all data should be considered),  
    plots the contig sizes from the biggest to the smallest and the N50 as horizontal line"""
    tmp = data
    title = "Sequence length"
    if lg != None:
        tmp = tmp[tmp["LG"] == lg]
        title += " of " + lg
        
    if tmp.shape[0] == 0:
        print("Warning cannot find information on {}".format(lg))
    else:
        n50 = computeN50(tmp, lg)
        print("N50: {}".format(n50))
        Y = list(tmp["Size"])
        Y.sort(reverse = True)
        X = list(range(1,len(Y) + 1))
        plt.title(title)
        plt.xlabel("# sequences")
        plt.ylabel("Size [bps]")
        plt.plot(X,Y)
        plt.hlines(n50, xmin=0, xmax=len(X) + 1, colors = ['red'], linestyles = 'dashed', label = "N50")
        plt.show()
        

if __name__ == "__main__":
    
    dataFile = "ctg_data.tsv"
    data = loadTSV(dataFile)

    
    computestats(data)
    

    n50 = computeN50(data)
    print("\nN50 of all data: {}".format(n50))

    n50 = computeN50(data, "LG1")
    print("N50 of LG1: {}".format(n50))
    
    plotN50(data)
    
    plotN50(data, "LG2")

    plotN50(data, "LG7")

    plotN50(data, "LG9")
    
    computeN50(data, "LG16")



````

````python
"""
NAME:
SURNAME:
Matricola: 

ReversibleQueue.py
"""

from MyQueue import MyQueue

from collections import deque

class ReversibleQueue(MyQueue):

        
    
    def printContent(self):
        """prints all the elements in the queue reporting the minimum and maximum value of the queue 
        using only standard queue operations. The queue must be unaffected by the method"""
        tmp = deque()
        m = self.top()
        M = self.top()
        cnt = 0
        while not self.isEmpty():
            val = self.dequeue()
            if val > M:
                M = val
            if val < m:
                m = val
            print("[{}]: {}".format(cnt,val))
            tmp.append(val)
            cnt += 1
        
        while len(tmp) > 0:
            self.enqueue(tmp.popleft())
        print("Min: {} Max:{}".format(m,M))
        

    
    def reverseK(self, K):
        """reverses the first K elements of the queue"""
        if K <= 0:
            return
        else:
            tmp = deque()
            cnt = 0
            while not self.isEmpty() and cnt < K:
                tmp.append(self.dequeue())
                cnt += 1
            while not self.isEmpty():
                tmp.appendleft(self.dequeue())
            
            while len(tmp) > 0:
                self.enqueue(tmp.pop())

        
        

if __name__ == "__main__":
    Q = ReversibleQueue()
    data = [ 72, 11, 2, 13, 87, 27, 44, 3, 1, -10, 4]
    for el in data:
        Q.enqueue(el)
    print("First in queue: {}".format(Q.top()))
    Q.printContent()

    print("Empty? {}".format(Q.isEmpty()))
    print("First in queue: {}".format(Q.top()))


    Q2 = ReversibleQueue()
    dataStr = ["APPLE", "BLUEBERRY", "PINEAPPLE"]
    for el in dataStr:
        Q2.enqueue(el)

    Q.reverseK(0)
    print("\nQ now (rev(0) unchanged):")
    Q.printContent()
    Q.reverseK(-5)
    print("\nQ now (rev(-5) unchanged):")
    Q.printContent()
    Q.reverseK(5)
    print("\nQ now (first 5 rev):")
    Q.printContent()
        
    print("\nQ2:")
    Q2.printContent()
    Q2.reverseK(0)
    print("\nQ2 now (rev(0) unchanged):")
    Q2.printContent()
    print("\nQ2 now (rev(rev(2))unchanged):")
    Q2.reverseK(2)
    Q2.reverseK(2)
    Q2.printContent()
    Q2.reverseK(10)
    print("\nQ2 now (rev(10):")
    Q2.printContent()
    
    
````
