````python
"""
NAME:
SURNAME:
Matricola: 

partA.py
"""
%reset -f 
import pandas as pd
from matplotlib import pyplot as plt
from Bio import SeqIO

def computeKmers(fasta_filename, K):
    """gets the name of the fasta file fasta-filename and an integer K specifying the k-mer size and returns 
    a pandas DataFrame with the following columns: SequenceID, SequenceLen, Kmer,  Start, End"""
    sequences_dict = SeqIO.to_dict(SeqIO.parse(fasta_filename, "fasta"))
    
    lengths = []
    seqIDS = []
    kmers = []
    starts = []
    ends = []
    for s in sequences_dict:
        sID = s
        #cast to str because it is actually a SeqRecord
        seq = str(sequences_dict[s].seq)
        L = len(seq)
        #note + 1 
        for i in range(0, len(seq)- K +1):
            kseq = seq[i:i+K]
            lengths.append(L)
            seqIDS.append(sID)
            kmers.append(kseq)
            starts.append(i+1)
            ends.append(i+K)
    data = pd.DataFrame({'SequenceID' : seqIDS, 
                         'SequenceLen' : lengths,
                         'Kmer' : kmers,
                         'Start' : starts,
                         'End' : ends
                        })
    return data


def showKmerStats(data): 
    """given the input data as returned from computeKmers, prints the total number of sequences and their ids, 
    total number of kmers, the total number of kmers per SequenceID and the 3 most frequent kmers of each sequence 
    """
    totSeq = data['SequenceID'].value_counts()
    print("The data contains info for the following {} sequences: ".format(len(totSeq)))
    for s in totSeq.index:
        print(s)
    
    print("\nThe total number of kmers identified is {}".format(data.shape[0]))
    for s in totSeq.index:
        print("{} kmers in sequence {}".format(data[data['SequenceID'] == s].shape[0],s))
    
    for s in totSeq.index:
        filt_data = data[data['SequenceID'] == s]
        kmer_cnt = filt_data['Kmer'].value_counts().head(3)
        print("The 3 most frequent kmers of {} are:".format(s))
        for kmer in kmer_cnt.index:
            print("\t{} present {} times".format(kmer,kmer_cnt[kmer]))

        


def plotKmer(data, seqID, kmer): 
    """given the input data as returned from computeKmers, a sequence identifier seqID and a kmer sequence kmer, 
    if the sequence seqID is present in the input data and contains the kmer sequence, a presence-absence plot with all 
    the positions where the kmer is located should be made, otherwise a warning message should be reported. 
    In the presence-absence plot, for all positions of the seqID sequence, a zero is reported if that position 
    is part of the kmer, 1 otherwise"""
    
    
    filt_data = data[data['SequenceID'] == seqID]
    if filt_data.shape[0] == 0:
        print("Warning: sequence {} is not present in the dataset".format(seqID))
    else:
        filt_data = filt_data[filt_data["Kmer"] == kmer]
        if filt_data.shape[0] == 0:
            print("Warning: kmer {} is not present in {}".format(kmer,seqID))
        else:
            seqL = filt_data['SequenceLen'].iloc[0]
            X = list(range(1,seqL+1))
            Y = [0 for x in range(len(X))]
            kmer_borders = filt_data[['Start', 'End']]
            for el in range(len(kmer_borders)):
                s = kmer_borders.iloc[el]['Start']
                e = kmer_borders.iloc[el]['End']
                
                #print(kmer_borders.iloc[el]['Start','End'])
                Y[s:e+1] = [1 for x in range(s,e+1)]
            plt.xlabel('Position [bps]')
            plt.ylabel('Presence/absence')
            plt.title("kmer {} on sequence {}".format(kmer,seqID))
            plt.plot(X,Y)
            plt.show()
                

    




dataFile = "../file_samples/june_sequences.fasta"
data = computeKmers(dataFile,5)
print("First 5 entries (k=5)")
print(data.head(5))

data = computeKmers(dataFile,7)
print("\nFirst 5 entries (k=7)")
print(data.head(5))
print("\nLast 5 entries (k=7)")
print(data.tail(5))

print("\n")

showKmerStats(data)


plotKmer(data, "Seq17e_pc1", "TATATAT")

plotKmer(data, "my_sequence", "TATATAT")

plotKmer(data, "Seq17e_pc1", "NNNNNNN")

plotKmer(data, "Seq17s_pc1", "ATTTTTG")


plotKmer(data, "Seq17e_pc1", "TTGAGGG")

````

````python
"""Exercise 2"""
%reset -f
from collections import deque


def F(n):
    tmp = deque()
    if  n<= 0:
        return 1
    if n < 5:
        return 1/n
    if n >= 5:
        mult = [1]
        tmp.append(1)
        for x in range(2,6):
            mult.append(x)
            tmp.appendleft(1/(x-1))
        
        res = -1
        for i in range(5,n+1):
            res = sum([mult[i]*tmp[i] for i in range(0,5)])
            tmp.pop()
            tmp.appendleft(res)
        return res

for i in range(0,12):
    print("F({}) = {:.3f}".format(i,F(i)))

print("F(125) = {}".format(F(125)))

        
````

````python
"""
NAME:
SURNAME:
Matricola: 

ExtendedStack.py
"""
from collections import deque
from MyStack import MyStack

class ExtendedStack(MyStack):

        
    
    def printContent(self):
        """prints all the elements in the stack reporting the minimum and maximum value of the stack 
        using only standard stack operations. The stack must be unaffected by the method"""
        tmpList = deque()
        m = self.peek()
        M = self.peek()
        print("Stack content: ")
        peekStr = " <-- peek()"
        while not self.isEmpty():
            tmpList.append(self.pop())
            print("\t{}{}".format(tmpList[-1], peekStr))
            peekStr = ""
            if tmpList[-1] > M:
                M = tmpList[-1]
            if tmpList[-1] < m:
                m = tmpList[-1]
        
        #put all elements back in
        while len(tmpList) > 0:
            self.push(tmpList.pop())
            
        print("Minimum is: {}".format(m))
        print("Maximum is: {}".format(M))
    
    def interleave(self, otherStack):
        """interleves the elements of self with those of other stack"""
        
        selfTmp = deque()
        while not self.isEmpty():
            selfTmp.append(self.pop())
        
        otherTmp = deque()
        while not otherStack.isEmpty():
            otherTmp.append(otherStack.pop())
        
        while len(selfTmp) > 0 or len(otherTmp) > 0:
            if len(selfTmp) > 0:
                self.push(selfTmp.pop())
            if len(otherTmp) > 0:
                self.push(otherTmp[-1])
                otherStack.push(otherTmp.pop())
        
        

if __name__ == "__main__":
    S = ExtendedStack()
    data = [ 72, 11, 2, 13, 87, 27, 44, 3, 1, -10, 4]
    for el in data:
        S.push(el)
    print("Top of the stack: {}".format(S.peek()))
    S.printContent()
    
    print("Empty? {}".format(S.isEmpty()))
    print("Top of the stack: {}".format(S.peek()))
    
    S1 = ExtendedStack()
    
    dataStr = ["elephant", "giraffe", "buffalo", "zebra", "horse"]
    for el in dataStr:
        S1.push(el)
    
    S2 = ExtendedStack()
    dataStr = ["APPLE", "BLUEBERRY", "PINEAPPLE"]
    for el in dataStr:
        S2.push(el)
    
    print("\nS1:")
    S1.printContent()
    print("\nS2:")
    S2.printContent()
    
    S1.interleave(S2)
    print("\nS1 now:")
    S1.printContent()
    print("\nS2 now:")
    S2.printContent()
    
    
````
